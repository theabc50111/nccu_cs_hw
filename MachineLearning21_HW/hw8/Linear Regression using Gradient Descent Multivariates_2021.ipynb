```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
# Preprocessing Input data
# 直接使用 local file
##### $ y = f(x1, x2) = 0.3 x1 - 0.2 x2 + 3.2 + 3(uniform[0,1]-0.5)
data = pd.read_csv('multivariate.csv')
```

```python
X1= data.iloc[:,0]
X2= data.iloc[:,1]
Y = data.iloc[:, 2]
```

```python
plt.scatter(X1, Y)
plt.show()
```

```python
plt.scatter(X2, Y)
plt.show()
```

```python
# Building the model
m1 = 5
m2 = 5
c = 5

L = 0.001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X1)) # Number of elements in X

%matplotlib inline

# Performing Gradient Descent 
for i in range(epochs): 
    plt.clf()
    Y_pred = m1*X1 + m2*X2 + c  # The current predicted value of Y
    D_m1 = (-2/n) * sum(X1 * (Y - Y_pred))  # Derivative wrt m1
    D_m2 = (-2/n) * sum(X2 * (Y - Y_pred))  # Derivative wrt m1
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m1 = m1 - L * D_m1  # Update m
    m2 = m2 - L * D_m2  # Update m
    c = c - L * D_c  # Update c
    Y_pred = m1*X1 + m2*X2 + c  # The current predicted value of Y

#    if (i%2==0):
#        plt.scatter(X1, Y)
#        plt.plot([min(X1), max(X1)], [min(Y_pred), max(Y_pred)], color='red') # predicted
#        plt.show()
    
    if (i%10==0):
        print ('iteration= {:3}, m1={:3.4}, m2={:3.4}, c={:3.4}'.format(i, m1, m2, c))
print ('iteration: {:3}, m1={:3.4}, m2={:3.4}, c={:3.4}'.format(i, m1, m2, c))
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np

X = np.concatenate((X1.values.reshape(-1,1), X2.values.reshape(-1,1)),axis=1)
model = linear_model.LinearRegression()
model.fit(X,Y.values)
model.coef_, model.intercept_
```
